```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Function to read data from a text file, extracting only 'Episode' and 'Episode_Reward'
def read_data_from_file(filename):
    # Reading the file and skipping any unnecessary columns
    df = pd.read_csv(filename, sep='\t', header=0)  # assuming tab-separated values
    return df[['Episode', 'Episode_Reward']]  # Select only relevant columns

# Read data from the text files for both Baseline DQN and Multistep Rewards
df_baseline = read_data_from_file('comparisons/base_breakout_v1.txt')
df_multistep = read_data_from_file('comparisons/msl_breakout_v1.txt')

# Smoothing function using a window size of 20
def smooth_rewards(rewards, window_size=500):
    return np.convolve(rewards, np.ones(window_size)/window_size, mode='valid')

# Smooth the rewards for both Baseline DQN and Multistep
smoothed_baseline = smooth_rewards(df_baseline['Episode_Reward'], window_size=500)
smoothed_multistep = smooth_rewards(df_multistep['Episode_Reward'], window_size=500)

# Adjust episode numbers for the smoothed data (since smoothing reduces length)
episode_baseline_smoothed = df_baseline['Episode'][500-1:].values
episode_multistep_smoothed = df_multistep['Episode'][500-1:].values

# Plotting the graph
plt.figure(figsize=(10, 6))

# Plot Baseline DQN Rewards (blue line)
plt.plot(episode_baseline_smoothed, smoothed_baseline, label="Baseline DQN Rewards", color='blue')

# Plot Multistep Learning Rewards (red line)
plt.plot(episode_multistep_smoothed, smoothed_multistep, label="Multistep Learning Rewards", color='red')

# Title and labels
plt.title("Comparison of episode rewards over time smoothed over windows size 20 (MinAtar/Breakout-v1)")
plt.xlabel("Episode Number")
plt.ylabel("Episode Reward Smoothed")

# Show legend
plt.legend()

# Show the plot
plt.show()
```
